### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
# Load in csv
city_df = pd.read_csv("../output_data/cleancities.csv")
city_df.describe()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Configure gmaps with API key
gmaps.configure(api_key=g_key)
```

```python
# Store 'Lat' and 'Lng' into  locations 
locations = city_df[["Lat", "Lng"]].astype(float)

# Convert Humidity to float and store
Humidity = city_df["Humidity"].astype(float)
```

```python
# Create a Heatmap layer
fig = gmaps.figure(center=(40.0, -5.0), zoom_level=8)

heat_layer = gmaps.heatmap_layer(locations, weights=Humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 1)

fig.add_layer(heat_layer)

fig
```

## Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# Filter cities to get vacation ctiies with temperatures between 18 to 35, wind speed less than equals to 4 and cloudiness less than equals to 13

filter_condition = ((city_df["Max Temp"] >= 18) & (city_df["Max Temp"] <= 35)) & (city_df["Wind Speed"] <= 4) & (city_df["Cloudiness"] <= 13)

vacation_city_df = city_df[filter_condition]
vacation_city_df = vacation_city_df.reset_index()
del vacation_city_df["index"]
vacation_city_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
radius = 5000
params = {
    "radius": radius,
    "type": "hotel",
     "key": g_key,
}
# use iterrows to iterate through pandas dataframe
# tuple unpacking
# parallel assignment
for index, row in vacation_city_df.iterrows():
    
    # get lat, lng from df
    lat = row["Lat"]
    lng = row["Lng"]

    # change location each iteration while leaving original params in place
    params["location"] = f"{lat},{lng}"

    # assemble url and make API request
    print(f"Retrieving Results for Index {index}: {row['City']}.")
    response = requests.get(base_url, params=params).json()
    
    # extract results
    results = response['results']
    
    try:
        print(f"First hotel is {results[0]['name']}.")
        
        vacation_city_df.loc[index, 'Hotel Name'] = results[0]['name']
       
    
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
    except Exception:  
        print("Exception" + Exception)
print("Completed Hotel retrieval info------------")
```

```python
vacation_city_df
hotel_df = vacation_city_df
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map

hotel_layer = gmaps.symbol_layer(
    locations, fill_color='rgba(0, 150, 0, 0.4)',
    stroke_color='rgba(0, 0, 150, 0.4)', scale=5,
    info_box_content=hotel_info
)


fig.add_layer(hotel_layer)
fig

# Display figure
```
